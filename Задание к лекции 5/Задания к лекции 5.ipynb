```python
# Задание 1

import csv
# Читаем файл и считаем доход
product_revenue = {}  

with open('data.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # пропускаем заголовки
    
    for row in reader:
        product_name = row[1] 
        quantity = int(row[2])
        price = int(row[3])   
        revenue = quantity * price 
        
        
        if product_name in product_revenue:
            product_revenue[product_name] += revenue
        else:
            product_revenue[product_name] = revenue

# Выводим результаты
print("Доход по каждому продукту:")
for product, revenue in product_revenue.items():
    print(f"{product}: {revenue} руб.")
```

```python
#Задание 2
import csv
import json

# Читаем файл и считаем доход
product_revenue = {} 

with open('data.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # пропускаем заголовки
    
    for row in reader:
        product_name = row[1] 
        quantity = int(row[2])
        price = int(row[3])   
        revenue = quantity * price 
        
        
        if product_name in product_revenue:
            product_revenue[product_name] += revenue
        else:
            product_revenue[product_name] = revenue

# Выводим результаты по доходам
print("Доход по каждому продукту:")
for product, revenue in product_revenue.items():
    print(f"{product}: {revenue} руб.")

#Сохраняем исходные данные в JSON
sales_data = []

# Читаем CSV файл для создания JSON
with open('data.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    headers = next(reader)  # получаем заголовки
    
    for row in reader:
        sale = {
            headers[0]: int(row[0]),    
            headers[1]: row[1],         
            headers[2]: int(row[2]),    
            headers[3]: int(row[3])     
        }
        sales_data.append(sale)

# Сохраняем в JSON файл
with open('data.json', 'w', encoding='utf-8') as file:
    json.dump(sales_data, file, ensure_ascii=False, indent=2)

print("\nДанные успешно сохранены в data.json")
```

```python
import requests

comments_url = 'https://jsonplaceholder.typicode.com/comments'
comments = requests.get(comments_url).json()


post_comments = {} 

for comment in comments:
    post_id = comment['postId']
    
    if post_id in post_comments:
        post_comments[post_id] += 1
    else:
        post_comments[post_id] = 1


post_id_max_comments = max(post_comments, key=post_comments.get)
max_comments_count = post_comments[post_id_max_comments]

# 4. Получаем информацию о посте
post_url = f'https://jsonplaceholder.typicode.com/posts/{post_id_max_comments}'
post = requests.get(post_url).json()

# 5. Выводим результат
print(f"Пост с максимальным количеством комментариев:")
print(f"ID поста: {post['id']}")
print(f"Заголовок: {post['title']}")
print(f"Количество комментариев: {max_comments_count}")
```
