```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# Загрузка изображения
img = Image.open("image.png") 
print(f"Исходный формат: {img.format}, Размер: {img.size}")

# Проверка на PNG и обработка прозрачности
if img.format == 'PNG' and img.mode == 'RGBA':
    # Создание белого фона в формате RGBA
    background = Image.new("RGBA", img.size, (255, 255, 255, 255))
    img = Image.alpha_composite(background, img).convert("RGB")
    print("Прозрачные пиксели заменены белыми.")

# Изменение размера изображения до 100x100 пикселей
resized_img = img.resize((100, 100))
# Конвертация в черно-белый формат
gray_img = resized_img.convert("L")
# Применение фильтра Гаусса для размытия
blurred_img = gray_img.filter(ImageFilter.GaussianBlur(5))

# Сохранение обработанных изображений
resized_img.save("resized_image.jpg")
gray_img.save("gray_image.jpg")
blurred_img.save("blurred_image.jpg")

# Отображение изображений
fig, axs = plt.subplots(1, 4, figsize=(15, 5))
axs[0].imshow(img)
axs[0].set_title("Оригинал")
axs[0].axis("off")

axs[1].imshow(resized_img)
axs[1].set_title("Изменённый размер")
axs[1].axis("off")

axs[2].imshow(gray_img, cmap="gray")
axs[2].set_title("Черно-белый")
axs[2].axis("off")

axs[3].imshow(blurred_img, cmap="gray")
axs[3].set_title("С размытием")
axs[3].axis("off")

plt.show()
```
