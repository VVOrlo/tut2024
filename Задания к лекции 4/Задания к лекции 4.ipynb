Задание 1
Написать программу, которая подсчитывает количество
вхождений каждого символа в заданной строке и выводит
результат в виде словаря.

```python
stroka=input('Введите в строку:')
stroka1=stroka.replace(' ','')
strLen=len(stroka1)
dictStr={}
for i in range(strLen):
    if stroka1[i] in dictStr:
        dictStr[stroka1[i]]=dictStr[stroka1[i]]+1
    else:
        dictStr[stroka1[i]]=1
print(dictStr)
```

Задание 2
Написать программу, которая создаёт множество из списка
чисел, введенных пользователем, и выводит его сначала в
неизменном виде, а затем удаляет дубликаты и выводит
результат без дублей.

```python
def remove_dup(string):
    str1=set()
    reslt=[]
    for char in string:
        if char not in str1:
            str1.add(char)
            reslt.append(char)
    return reslt

stroka=input('Введите в строку:')
stroka1=list(stroka)
stroka2=remove_dup(stroka)
print(stroka1)
print(stroka2)
```
